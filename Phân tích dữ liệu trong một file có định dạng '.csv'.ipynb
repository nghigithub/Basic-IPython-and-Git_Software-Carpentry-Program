Vẽ đồ thị

```python
fig=plt.figure(figsize=(10,3)) #tạo figure với kích thước 10,3

#tạo các figure nhỏ hơn trong figure lớn vừa tạo
fig1=fig.add_subplot(1,3,1)
fig2=fig.add_subplot(1,3,2)
fig3=fig.add_subplot(1,3,3)

fig1.set_ylabel('avarage') #đặt tên trục y
fig1.plot(np.mean(data,axis=0)) #vẽ giá trị mean theo từng cột trên figure1

fig2.set_ylabel('max') 
fig2.plot(data.max(axis=0))

fig3.set_ylabel('min') 
fig3.plot(data.min(axis=0))

fig.tight_layout() #Kéo các hình cho vừa với kích thước 10,3 đã tạo

plt.show()
```

Đặt tỉ lệ cho các trục đồ thị

```python
fig=plt.figure(figsize=(10,3)) #tạo figure với kích thước 10,3

#tạo các figure nhỏ hơn trong figure lớn vừa tạo
fig1=fig.add_subplot(1,3,1)
fig2=fig.add_subplot(1,3,2)
fig3=fig.add_subplot(1,3,3)

fig1.set_ylabel('avarage') #đặt tên trục y
fig1.plot(np.mean(data,axis=0)) #vẽ giá trị mean theo từng cột trên figure1

fig2.set_ylabel('max') 
fig2.plot(data.max(axis=0))

fig3.set_ylabel('min') 
fig3.plot(data.min(axis=0))
fig3.set_ylim(0,6)

fig.tight_layout() #Kéo các hình cho vừa với kích thước 10,3 đã tạo

plt.show()
```

```python
fig=plt.figure(figsize=(10,3)) #tạo figure với kích thước 10,3

#tạo các figure nhỏ hơn trong figure lớn vừa tạo
fig1=fig.add_subplot(1,3,1)
fig2=fig.add_subplot(1,3,2)
fig3=fig.add_subplot(1,3,3)

fig1.set_ylabel('avarage') #đặt tên trục y
fig1.plot(np.mean(data,axis=0)) #vẽ giá trị mean theo từng cột trên figure1

fig2.set_ylabel('max') 
fig2.plot(data.max(axis=0))

fig3.set_ylabel('min') 
fig3.plot(data.min(axis=0))
fig3.set_ylim(np.min(data.min(axis=0)),np.max(data.min(axis=0))*1.1)

fig.tight_layout() #Kéo các hình cho vừa với kích thước 10,3 đã tạo

plt.show()
```

```python
fig=plt.figure(figsize=(3,10)) #tạo figure với kích thước 10,3

#tạo các figure nhỏ hơn trong figure lớn vừa tạo
fig1=fig.add_subplot(3,1,1)
fig2=fig.add_subplot(3,1,2)
fig3=fig.add_subplot(3,1,3)

fig1.set_xlabel('avarage') #đặt tên trục y
fig1.plot(np.mean(data,axis=0)) #vẽ giá trị mean theo từng cột trên figure1

fig2.set_xlabel('max') 
fig2.plot(data.max(axis=0))

fig3.set_xlabel('min') 
fig3.plot(data.min(axis=0))
fig3.set_ylim(np.min(data.min(axis=0)),np.max(data.min(axis=0))*1.1)

fig.tight_layout() #Kéo các hình cho vừa với kích thước 3,10 đã tạo

plt.show()
```

Kiểm tra dữ liệu trong file sử dụng hàm if

```python
data = np.loadtxt(fname='inflammation-03.csv', delimiter=',')
if np.max(data, axis=0)[0] == 0 and np.max(data, axis=0)[20] == 20:
    print('Suspicious looking maxima!')
elif np.sum(np.min(data, axis=0)) == 0:
    print('Minima add up to zero!')
else:
    print('Seems OK!')
```
